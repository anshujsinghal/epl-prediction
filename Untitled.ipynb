```python
def get_goals_scored(playing_stat):
    teams = {team: [] for team in playing_stat['HomeTeam'].unique()}
    
    for _, row in playing_stat.iterrows():
        teams[row.HomeTeam].append(row.FTHG)
        teams[row.AwayTeam].append(row.FTAG)

    GoalsScored = pd.DataFrame(teams).T
    GoalsScored.columns = range(1, GoalsScored.shape[1] + 1)
    
    # cumulative sum
    GoalsScored = GoalsScored.cumsum(axis=1)
    
    # shift so current matchweek excludes current match
    GoalsScored = GoalsScored.shift(axis=1).fillna(0)
    
    return GoalsScored


def get_goals_conceded(playing_stat):
    teams = {team: [] for team in playing_stat['HomeTeam'].unique()}
    
    for _, row in playing_stat.iterrows():
        teams[row.HomeTeam].append(row.FTAG)
        teams[row.AwayTeam].append(row.FTHG)

    GoalsConceded = pd.DataFrame(teams).T
    GoalsConceded.columns = range(1, GoalsConceded.shape[1] + 1)
    
    GoalsConceded = GoalsConceded.cumsum(axis=1)
    GoalsConceded = GoalsConceded.shift(axis=1).fillna(0)
    
    return GoalsConceded


def get_gss(playing_stat):
    GC = get_goals_conceded(playing_stat)
    GS = get_goals_scored(playing_stat)

    HTGS, ATGS, HTGC, ATGC = [], [], [], []
    
    matchweek = 0
    matches_per_week = playing_stat.groupby('Date').ngroups // 38 if 'Date' in playing_stat else 10

    for i, row in playing_stat.iterrows():
        ht, at = row.HomeTeam, row.AwayTeam
        
        HTGS.append(GS.loc[ht][matchweek])
        ATGS.append(GS.loc[at][matchweek])
        HTGC.append(GC.loc[ht][matchweek])
        ATGC.append(GC.loc[at][matchweek])
        
        if (i + 1) % matches_per_week == 0:
            matchweek += 1

    playing_stat['HTGS'] = HTGS
    playing_stat['ATGS'] = ATGS
    playing_stat['HTGC'] = HTGC
    playing_stat['ATGC'] = ATGC

    return playing_stat
```
